```python

select_countires_confirmed = time_series_confirmed[time_series_confirmed['Country/Region'] == SELECTED_COUNTRY]['3/21/20'].sum()
select_countires_recovered = time_series_recovered[time_series_recovered['Country/Region'] == SELECTED_COUNTRY]['3/21/20'].sum()
select_countires_deaths = time_series_deaths[time_series_deaths['Country/Region'] == SELECTED_COUNTRY]['3/21/20'].sum()
# mortality rate
select_countires_mortality_rate = (select_countires_deaths / select_countires_confirmed)*100
# recovery rate
select_countires_recovery_rate = (select_countires_recovered / select_countires_confirmed)*100
select_countires_mortality_rate, select_countires_recovery_rate
```

```python
# Time Series of Coronna data for the selected country
dates = list(time_series_confirmed.columns)[4:]
series_select_countires_confirmed = []
series_select_countires_recovered = []
series_select_countires_deaths = []
for i in dates:
    series_select_countires_confirmed.append(time_series_confirmed[time_series_confirmed['Country/Region'] == SELECTED_COUNTRY][i].sum())
    series_select_countires_recovered.append(time_series_recovered[time_series_recovered['Country/Region'] == SELECTED_COUNTRY][i].sum())
    series_select_countires_deaths.append(time_series_deaths[time_series_deaths['Country/Region'] == SELECTED_COUNTRY][i].sum())
    series_mortality_rate = [(x/y)*100 for x, y in zip(series_select_countires_deaths, series_select_countires_confirmed)]
    series_recovery_rate = [(x/y)*100 for x, y in zip(series_select_countires_recovered, series_select_countires_confirmed)]
    
df_recovery = pd.DataFrame(columns = ['Dates','RecoveryRate','MoratalityRate'])
df_recovery['Dates'] = pd.Series(dates)
df_recovery['RecoveryRate'] = pd.Series(series_recovery_rate)
df_recovery['MoratalityRate'] = pd.Series(series_mortality_rate)
```

```python
COUNTRIES = list(set(daily_report['Country/Region']))
series_select_countires_confirmed = []
series_select_countires_recovered = []
series_select_countires_deaths = []
for i in COUNTRIES:
    series_select_countires_confirmed.append(time_series_confirmed[time_series_confirmed['Country/Region'] == i]['3/21/20'].sum())
    series_select_countires_recovered.append(time_series_recovered[time_series_recovered['Country/Region'] == i]['3/21/20'].sum())
    series_select_countires_deaths.append(time_series_deaths[time_series_deaths['Country/Region'] == i]['3/21/20'].sum())
series_mortality_rate = [(x/y)*100 for x, y in zip(series_select_countires_deaths, series_select_countires_confirmed)]
series_recovery_rate = [(x/y)*100 for x, y in zip(series_select_countires_recovered, series_select_countires_confirmed)]
df_country_stats = pd.DataFrame(columns = ['Countries','Confirmed','Deaths','Recovered','MortalityRate','RecoveryRate'])
df_country_stats['Countries'] = pd.Series(COUNTRIES)
df_country_stats['Confirmed'] = pd.Series(series_select_countires_confirmed)
df_country_stats['Deaths'] = pd.Series(series_select_countires_deaths)
df_country_stats['Recovered'] = pd.Series(series_select_countires_recovered)
df_country_stats['MortalityRate'] = pd.Series(series_mortality_rate)
df_country_stats['RecoveryRate'] = pd.Series(series_recovery_rate)
```
